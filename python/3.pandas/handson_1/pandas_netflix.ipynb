### split columns

```python
# quero que seja feito split sempre que for encontrado uma ', ' e nisto vai criar um column ao lado

netflix["date_added"].str.split(", ",expand=True)
```

```python
# aqui estoua usar o dataframe e criar duas colunas ,  1 para ter dadosd e de Dia e mes e a outra coluna para ter dados de ano
netflix[["date part1","date part2"]] = netflix["date_added"].str.split(', ', expand=True)
netflix
```

```python
# rename method

netflix.rename(columns={"date part1": "month_day"},  inplace=True)
netflix.rename(columns={"date part2": "Year"},  inplace=True)
netflix
```

```python
# existem caso onte posso precisar de substituir os NAN para alguma mais significativo

netflix["cast"].fillna(value="no cast", inplace=True)
netflix
```

```python
# aqui eu vou ter os elementos da coluna listed_in numa forma de list,  e se usar o len(),  netse mesmo formato vou saber quantos elementos tem dentro dessa lista
#netflix["genre_count"] = netflix["listed_in"].map(lambda x: x.split(","))

# isto serve para saber o len de elementos dentro de uma lista
netflix["genre_count"] = netflix["listed_in"].map(lambda x: len(x.split(",")))
netflix
```

```python
netflix["non_sensational_columns"] = netflix.apply(lambda x: len(x["cast"].split(","))+ len(x["listed_in"].split(",")), axis=1)
netflix
```

### group by and aggregations  

```python
# vamos agrupar por pais e fazer o count por show_id
netflix.groupby("country")["show_id"].count()

# podemos ainda melhorar o formato dos daods
netflix.groupby("country")["show_id"].count().reset_index()
```

```python
# criar um varaivel apenas para ter sort dos paises com mais shoes no netflix
netflix_movies_by_country = netflix.groupby("country")["show_id"].count().reset_index()
# com base na variavel que criamos anteriormente vamos apenas,  usar um sort_values indicar qual é a coluna que vai basear o nosso sort // por default o sort é sempre ascending
# neste caso queremos ver os paises que tem o maior numero de shows
netflix_movies_by_country.sort_values(by="show_id", ascending=False)
```

```python
# a mesma linha cell anterior posso colocar tudo na mesma linha se quiser e temos o mesmo output,  apenas tenho de chamar pela variabel.

# o pandas permite que consigamos usar mais do que um method seguido,  isto é bom se pensarmos numa prespectiva de transformaçao
netflix_movies_by_country = netflix.groupby("country")["show_id"].count().reset_index().sort_values(by="show_id", ascending=False)
netflix_movies_by_country
```

### joins and unions

```python
second_dataset = pd.read_csv(pwd +"/netflix_titles_second.csv")
second_dataset
```

### pivot and melt

```python
'''
este pode ser perfeitamente o standard para abordar os jois de 2 dataframes

netflix_merged = pd.merge(left=, right=,  how=, left_on=,right_om= )

vamos ter de ter 2 dataset:
> new_dataset
> netflix_movies_by_country
> vamos fazer um inner join
> depois indicar qual o campo que vamos fazer o join
'''

netflix_merged = pd.merge(left=new_dataset, right=netflix_movies_by_country,  how="inner", left_on=["country"],right_on=["country"])
netflix_merged
```

```python
# pivot a table

'''
> index,  é o ponto de valiçao quero ver por paises
> quero que a colunas sejam todos os elementos que estejam presentes em type (se é um movie ou um tv show)
> quero se conte o numero de titles por type,  isto vai fazer um count,  com base nos countries

> depois do count temos,  fill_value=0 , que é basicamente qualquer paises que nao tenha um match ou umcount nestes 2 types apresenta 0

nota:  sempre que precisamos de dar um limpeza nos dados,  usamos sempre no fim .reset_index()

'''


pivot_table = netflix.pivot_table(index="country",  columns="type", values = "title", aggfunc="count", fill_value=0).reset_index()
pivot_table
```
